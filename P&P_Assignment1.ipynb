```python
file_path = '/content/drive/MyDrive/SkillStorm Files/Pride_and_Prejudice.txt'


myfile = open(file_path, "r", encoding="utf8")
raw_text = myfile.readlines()[36:14229]
print(raw_text)
```

```python
formatted = ""

for clean in raw_text:
  if clean!= '\n':
    clean = clean.strip()
    formatted = formatted + str(clean) + " "
formatted = formatted.replace('.', '')
print(formatted)
```

```python
countable = formatted.split()
print(str(len(countable)))
```
